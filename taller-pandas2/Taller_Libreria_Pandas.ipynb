# Punto 2 – Operaciones Matemáticas entre Columnas

En esta sección se realizarán diferentes operaciones matemáticas utilizando las columnas del DataFrame. Se crearán nuevas variables derivadas que permitirán analizar el comportamiento de las ventas y los costos asociados a cada producto.

## Total de ventas por producto

Se crea una nueva columna llamada Total_Venta, calculada multiplicando el precio de cada producto por la cantidad vendida.

```python
# Crear columna Total_Venta
df["Total_Venta"] = df["Precio"] * df["Cantidad"]

# Mostrar resultados
df[["Producto", "Precio", "Cantidad", "Total_Venta"]]
```

## Costo total por producto

Se crea la columna Costo_Total sumando el costo de envío al total de ventas de cada producto.

```python
# Crear columna Costo_Total
df["Costo_Total"] = df["Total_Venta"] + df["Costo_Envio"]

# Mostrar resultados
df[["Producto", "Total_Venta", "Costo_Envio", "Costo_Total"]]
```

## Promedio de ventas

Se calcula el valor promedio de la columna Total_Venta para conocer el comportamiento general de las ventas.

```python
# Calcular promedio
promedio_ventas = df["Total_Venta"].mean()

print(f"Promedio de Total_Venta: ${promedio_ventas:,.2f}")
```

## Producto con mayor venta

Se identifica el producto que generó el mayor valor de ventas dentro del conjunto de datos.

```python
# Producto con mayor Total_Venta
producto_mayor = df.loc[df["Total_Venta"].idxmax()]

producto_mayor
```

### Análisis

El producto mostrado anteriormente corresponde al artículo con el mayor valor de ventas, considerando el precio unitario y la cantidad registrada.

## Producto con menor venta

Se identifica el producto que registró el menor valor de ventas dentro del conjunto de datos.

```python
# Producto con menor Total_Venta
producto_menor = df.loc[df["Total_Venta"].idxmin()]

producto_menor
```

### Análisis

El producto mostrado anteriormente corresponde al artículo con el menor valor de ventas dentro del conjunto de datos analizado.

# Punto 3 – Porcentajes y Análisis

En esta sección se calcularán diferentes indicadores porcentuales a partir del Total_Venta. Estos indicadores permiten estimar impuestos, ganancias, pérdidas y la participación de cada producto en el total de ventas.

## Cálculo del IVA

Se crea una nueva columna llamada IVA que representa el 19% del valor de Total_Venta para cada producto.

```python
# Calcular IVA (19%)
df["IVA"] = df["Total_Venta"] * 0.19

# Mostrar resultados
df[["Producto", "Total_Venta", "IVA"]].head()
```

## Ganancia estimada

Se calcula una ganancia estimada equivalente al 25% del Total_Venta para cada producto.

```python
# Calcular ganancia estimada (25%)
df["Ganancia_Estimada"] = df["Total_Venta"] * 0.25

# Mostrar resultados
df[["Producto", "Total_Venta", "Ganancia_Estimada"]].head()
```

## Pérdida estimada

Se calcula una pérdida estimada equivalente al 0.05% del Total_Venta para cada producto.

```python
# Calcular pérdida estimada (0.05%)
df["Perdida_Estimada"] = df["Total_Venta"] * 0.0005

# Mostrar resultados
df[["Producto", "Total_Venta", "Perdida_Estimada"]].head()
```

## Participación porcentual de las ventas

Se calcula qué porcentaje representa cada producto respecto al total general de ventas del conjunto de datos.

```python
# Calcular total general de ventas
total_general_ventas = df["Total_Venta"].sum()

# Crear columna de porcentaje
df["Porcentaje_Ventas"] = (
    df["Total_Venta"] / total_general_ventas
) * 100

# Mostrar resultados
df[["Producto", "Total_Venta", "Porcentaje_Ventas"]].head()
```

## Productos con más del 15% de participación

Se filtran los productos cuya participación en las ventas totales es superior al 15%.

```python
# Filtrar productos con participación superior al 15%
productos_destacados = df[df["Porcentaje_Ventas"] > 15]

productos_destacados
```

### Análisis

Los productos mostrados anteriormente representan más del 15% de las ventas totales. Esto indica que tienen una participación significativa dentro del conjunto de datos y pueden considerarse productos de alto impacto en los ingresos.

# Punto 4 – Estadística Descriptiva

En esta sección se calculan diferentes medidas estadísticas que permiten comprender mejor el comportamiento de los datos. Estas medidas ayudan a identificar tendencias centrales, dispersión y valores extremos dentro del conjunto de datos.

## Media del Precio

La media representa el valor promedio de los precios de todos los productos registrados en el conjunto de datos.

```python
# Calcular media del precio
media_precio = df["Precio"].mean()

print(f"Media del Precio: ${media_precio:,.2f}")
```

### Interpretación

La media indica el precio promedio de los productos incluidos en el análisis. Permite tener una visión general del nivel de precios manejado por la empresa.

## Mediana del Precio

La mediana corresponde al valor central cuando los precios se ordenan de menor a mayor.

```python
# Calcular mediana
mediana_precio = df["Precio"].median()

print(f"Mediana del Precio: ${mediana_precio:,.2f}")
```

### Interpretación

La mediana divide el conjunto de datos en dos partes iguales. Es útil porque no se ve tan afectada por precios extremadamente altos o bajos.

## Moda del Precio

La moda representa el precio que más veces se repite dentro del conjunto de datos.

```python
# Calcular moda
moda_precio = df["Precio"].mode()

print("Moda del Precio:")
print(moda_precio)
```

### Interpretación

La moda permite identificar el precio más frecuente entre los productos analizados. Si aparecen varios valores, significa que existe más de un precio con la misma frecuencia máxima.

## Desviación estándar de la Cantidad

La desviación estándar mide qué tan dispersas están las cantidades respecto al promedio.

```python
# Calcular desviación estándar
desviacion_cantidad = df["Cantidad"].std()

print(f"Desviación estándar de Cantidad: {desviacion_cantidad:.2f}")
```

### Interpretación

Una desviación estándar alta indica que las cantidades varían considerablemente entre productos. Una desviación baja indica que las cantidades son más homogéneas.

## Valor máximo de Total_Venta

Se identifica el mayor valor de ventas registrado en el conjunto de datos.

```python
# Valor máximo
max_venta = df["Total_Venta"].max()

print(f"Valor máximo de Total_Venta: ${max_venta:,.2f}")
```

### Interpretación

Este valor corresponde a la venta más alta registrada entre todos los productos analizados.

## Valor mínimo de Total_Venta

Se identifica el menor valor de ventas registrado en el conjunto de datos.

```python
# Valor mínimo
min_venta = df["Total_Venta"].min()

print(f"Valor mínimo de Total_Venta: ${min_venta:,.2f}")
```

### Interpretación

Este valor corresponde a la venta más baja registrada dentro del conjunto de datos.

## Resumen de resultados

A continuación se presenta un resumen consolidado de las medidas estadísticas calculadas.

```python
resumen_estadistico = pd.DataFrame({
    "Medida": [
        "Media Precio",
        "Mediana Precio",
        "Desviación Estándar Cantidad",
        "Máximo Total_Venta",
        "Mínimo Total_Venta"
    ],
    "Valor": [
        media_precio,
        mediana_precio,
        desviacion_cantidad,
        max_venta,
        min_venta
    ]
})

resumen_estadistico
```

# Punto 5 – Filtrado y Ordenamiento

En esta sección se aplican filtros sobre el conjunto de datos para obtener información específica y se realiza un ordenamiento según el valor de las ventas.

## Productos de la categoría Tecnología

Se filtran los registros pertenecientes a la categoría Tecnología para analizar únicamente los productos tecnológicos.

```python
# Filtrar productos de la categoría Tecnología
productos_tecnologia = df[df["Categoria"] == "Tecnología"]

productos_tecnologia
```

```python
# Ordenar de mayor a menor según Total_Venta
productos_ordenados = df.sort_values(
    by="Total_Venta",
    ascending=False
)

productos_ordenados
```
