## **3.1 NumPy**

### ¿Qué es NumPy?

**NumPy** (*Numerical Python*) es una librería fundamental de Python especializada en la computación científica y el procesamiento eficiente de datos numéricos. Su principal función es proporcionar estructuras de datos multidimensionales y herramientas optimizadas para realizar operaciones matemáticas complejas de manera rápida y eficiente.

### ¿Por qué NumPy es más eficiente que las listas tradicionales de Python?

NumPy es más eficiente que las listas tradicionales de Python debido a que almacena los datos en bloques continuos de memoria y utiliza código optimizado escrito en lenguaje C. Esto permite ejecutar operaciones matemáticas de forma mucho más rápida, reducir el consumo de recursos y evitar el uso excesivo de bucles `for`.

Sus principales ventajas son:

* **Almacenamiento contiguo:** permite acceder a los datos de forma más rápida y eficiente.
* **Vectorización:** realiza operaciones sobre grandes conjuntos de datos simultáneamente.
* **Optimización:** utiliza código interno altamente eficiente para mejorar el rendimiento.

### ¿Qué es un ndarray?

El **ndarray** (*N-dimensional Array*) es la estructura de datos principal de NumPy. Se trata de un arreglo multidimensional homogéneo, es decir, todos sus elementos deben ser del mismo tipo de dato. Esta estructura permite almacenar y manipular grandes volúmenes de información de forma eficiente.

### ¿Qué relación existe entre NumPy y el álgebra lineal?

NumPy incorpora herramientas de álgebra lineal que permiten realizar operaciones matemáticas fundamentales como la manipulación de vectores y matrices, el cálculo de determinantes, matrices inversas, transpuestas y multiplicaciones matriciales. Estas operaciones constituyen la base matemática utilizada en Inteligencia Artificial y Machine Learning.

### ¿Cómo utilizan los modelos de Deep Learning las matrices de NumPy?

Los modelos de Deep Learning utilizan matrices para representar los datos de entrada, los pesos y los sesgos de las redes neuronales. Durante el entrenamiento, se realizan millones de operaciones matemáticas entre estas matrices para que el modelo aprenda patrones y pueda realizar predicciones de manera automática.

## **4.3 Generación de datos aleatorios**

Complete el código, ejecútelo y grafique la distribución norma



```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generar 1000 datos con distribución normal
datos = np.random.normal(
    loc=50,
    scale=10,
    size=1000
)

# Graficar la distribución
sns.histplot(datos, kde=True)

plt.title("Distribución Normal")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.show()
```

**Interpretación**

Al observar la gráfica, se observa que la mayor parte de los datos se encuentra alrededor del valor 50, que corresponde a la media de la distribución. A medida que los valores se alejan de este punto, la cantidad de datos disminuye, por lo que los valores muy bajos o muy altos aparecen con menor frecuencia. Esto genera la característica forma de campana de una distribución normal.

###4.3.1. ¿Qué significa distribución normal?

La distribución normal es una forma de representar cómo se distribuyen los datos alrededor de un valor central. Se caracteriza por tener una forma de campana simétrica, donde la mayoría de los valores se concentra cerca de la media y los valores más alejados aparecen con menor frecuencia. Este tipo de distribución es muy común en fenómenos naturales, sociales y científicos.

###4.3.2. ¿Por qué muchos algoritmos asumen distribuciones aproximadamente normales?

Muchos algoritmos asumen distribuciones aproximadamente normales porque este tipo de datos facilita el análisis estadístico y mejora el rendimiento de los modelos. Al estar la información concentrada alrededor de un valor central y existir menos valores extremos, es más sencillo identificar patrones, realizar predicciones y obtener resultados más estables y precisos.

# 7. Generación de datos sintéticos

Para simular el comportamiento de una fábrica inteligente, se generarán 500 observaciones sintéticas que representarán el estado operativo de distintas máquinas industriales.

Los datos no serán completamente aleatorios, sino que se establecerán relaciones lógicas entre las variables para que el análisis exploratorio produzca resultados coherentes y representativos de un entorno industrial real.

Se asume que un incremento en la temperatura, la vibración, las horas de uso y el consumo eléctrico aumenta la probabilidad de que una máquina presente una falla.

```python
# Número de observaciones
n = 500

# Generación de variables

temperatura = np.random.normal(75, 10, n)

vibracion = np.random.normal(4, 1.2, n)

horas_uso = np.random.randint(500, 5000, n)

consumo_electrico = np.random.normal(120, 20, n)

# Regla para determinar la falla

puntaje_falla = (
    (temperatura > 85).astype(int)
    + (vibracion > 5).astype(int)
    + (horas_uso > 3500).astype(int)
    + (consumo_electrico > 140).astype(int)
)

falla = np.where(puntaje_falla >= 2, "Sí", "No")

# Construcción del DataFrame

datos = pd.DataFrame({
    "Temperatura del Motor": temperatura,
    "Vibración": vibracion,
    "Horas de Uso": horas_uso,
    "Consumo Eléctrico": consumo_electrico,
    "Falla": falla
})

# Mostrar las primeras filas

datos.head(15)
```

---

# 8. Análisis Estadístico (NumPy)

En esta fase, se utilizará **NumPy** para extraer estadísticas descriptivas fundamentales de las variables numéricas simuladas. Esto nos permitirá comprender la tendencia central, la dispersión y la distribución de los datos operativos de las máquinas antes de proceder a la visualización.

Se calcularán métricas como la media, mediana, desviación estándar, varianza y percentiles.

```python
# Lista de columnas numéricas para iterar
columnas_numericas = ["Temperatura del Motor", "Vibración", "Horas de Uso", "Consumo Eléctrico"]

print("="*50)
print(" ESTADÍSTICAS DESCRIPTIVAS DE LAS MÁQUINAS")
print("="*50)

for col in columnas_numericas:
    # Convertir la columna de Pandas a un array de NumPy
    data_array = datos[col].values

    # Cálculos estadísticos con NumPy
    media = np.mean(data_array)
    mediana = np.median(data_array)
    varianza = np.var(data_array)
    desviacion = np.std(data_array)
    p25, p50, p75 = np.percentile(data_array, [25, 50, 75])

    print(f"\n⚙️ Variable: {col.upper()}")
    print(f"   • Media:               {media:.2f}")
    print(f"   • Mediana:             {mediana:.2f}")
    print(f"   • Varianza:            {varianza:.2f}")
    print(f"   • Desviación Estándar: {desviacion:.2f}")
    print(f"   • Rango Intercuartil:  {p25:.2f} (P25) - {p75:.2f} (P75)")

print("\n" + "="*50)
print(" CONTEO DE FALLAS (Variable Objetivo)")
print("="*50)
print(datos['Falla'].value_counts())
```

# 9. Análisis Visual (Matplotlib)

A continuación, se empleará **Matplotlib** para generar representaciones gráficas básicas que faciliten la identificación visual de la distribución de los datos y posibles valores atípicos (*outliers*).

Se graficarán histogramas para comprender el comportamiento individual de las variables de temperatura y consumo, y gráficos de dispersión para observar correlaciones iniciales.

En esta sección se analiza de forma individual el comportamiento de las variables operativas clave utilizando histogramas y gráficos de dispersión independientes.

```python
# 1. Histograma de Temperatura
plt.figure(figsize=(6, 4))
plt.hist(datos['Temperatura del Motor'], bins=20, color='coral', edgecolor='black')
plt.title("Distribución de Temperatura")
plt.xlabel("Temperatura (°C)")
plt.ylabel("Frecuencia")
plt.show()

# 2. Histograma de Vibración
plt.figure(figsize=(6, 4))
plt.hist(datos['Vibración'], bins=20, color='mediumpurple', edgecolor='black')
plt.title("Distribución de Vibración")
plt.xlabel("Vibración (mm/s)")
plt.ylabel("Frecuencia")
plt.show()

# 3. Dispersión: Temperatura vs Vibración
plt.figure(figsize=(6, 4))
plt.scatter(datos['Temperatura del Motor'], datos['Vibración'], alpha=0.6, color='dodgerblue')
plt.title("Temperatura vs Vibración")
plt.xlabel("Temperatura (°C)")
plt.ylabel("Vibración (mm/s)")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# 4. Dispersión: Horas de Uso vs Consumo Eléctrico
plt.figure(figsize=(6, 4))
plt.scatter(datos['Horas de Uso'], datos['Consumo Eléctrico'], alpha=0.6, color='seagreen')
plt.title("Horas de Uso vs Consumo")
plt.xlabel("Horas de Uso")
plt.ylabel("Consumo (kWh)")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

### **Interpretación de los Gráficos de Matplotlib**

* **Distribución de Temperatura:** La mayoría de las temperaturas de los motores se concentran cerca de los 80°C. Las temperaturas superiores a los 90°C representan un riesgo severo de falla por sobrecalentamiento.
* **Distribución de Vibración:** Se observa una concentración típica alrededor de los 4 mm/s. Valores superiores a los 5.5 mm/s sugieren la presencia de desajustes mecánicos en el equipo.
* **Temperatura vs Vibración:** Se evidencia que las máquinas con mayores temperaturas de operación tienden a presentar mayores niveles de vibración, aumentando la probabilidad de falla del motor.
* **Horas de Uso vs Consumo:** A medida que aumentan las horas de operación acumuladas, las máquinas presentan un consumo energético más elevado debido al desgaste natural y la pérdida de eficiencia.

# 10. Visualización Avanzada y Multivariada (Seaborn)

**Seaborn** permite llevar el análisis un paso más allá al integrar paletas de colores basadas en categorías y cálculos estadísticos automáticos.

En esta etapa se construirán:
1. **Mapa de Calor (Heatmap):** Para identificar qué variables están fuertemente correlacionadas.

# 15. Reflexión de IA: Automatización de Toma de Decisiones en Mantenimiento Predictivo

### **Pregunta:** ¿Cómo podrían utilizarse técnicas de Machine Learning o Inteligencia Artificial para automatizar la toma de decisiones en este caso?

Para automatizar la toma de decisiones en la planta de producción basada en el análisis de este caso de estudio, se propone la implementación de un **Sistema de Lazo Cerrado (Closed-Loop) apoyado en Machine Learning Supervisado y No Supervisado**.

El proceso de automatización operaría bajo la siguiente arquitectura tecnológica:

#### **1. Detección Temprana y Predicción de Fallas (Modelamiento de Machine Learning)**
* **Modelos de Clasificación Supervisada:** Utilizando algoritmos como **XGBoost, Random Forest o Redes Neuronales Artificiales (ANN)**, entrenados con el histórico de datos recolectados (como el dataset sintético que analizamos). El modelo recibirá datos en tiempo real de los sensores y calculará constantemente la *Probabilidad de Falla ($P_{falla}$)* de cada máquina en una ventana de tiempo futura (por ejemplo, las próximas 48 horas de operación).
* **Modelos de Detección de Anomalías (No Supervisado):** Implementando algoritmos como **Isolation Forests o Autoencoders (Deep Learning)** para detectar patrones inusuales en la vibración o temperatura que no hayan sido registrados previamente en el histórico, previniendo fallas inéditas.

#### **2. Automatización de la Toma de Decisiones (Sin Intervención Humana Directa)**
En lugar de depender de que un operario humano observe una pantalla y tome una decisión manual, la Inteligencia Artificial automatizará las acciones de la siguiente manera:

* **Nivel 1: Alerta y Diagnóstico Automático (Notificación):** Si $P_{falla}$ supera el **60%**, la IA genera un diagnóstico predictivo automático (ej. *"Motor 4 presenta alta probabilidad de falla por sobrecalentamiento y desalineación mecánica"*) y envía una alerta instantánea vía bot de Telegram/Slack o correo al supervisor de turno.
* **Nivel 2: Generación Autónoma de Órdenes de Trabajo (Integración ERP):** Si $P_{falla}$ supera el **75%**, el modelo de IA se integra mediante APIs con el software de gestión de activos de la empresa (como SAP PM o Máximo). El sistema generará automáticamente una **Orden de Trabajo Preventiva**, reservará los repuestos necesarios en el inventario de la fábrica y asignará la tarea al técnico calificado disponible según su agenda.
* **Nivel 3: Control Adaptativo de Seguridad (Parada de Emergencia / Mitigación):** Si $P_{falla}$ supera el **90%** y los sensores detectan que la temperatura sobrepasa el umbral crítico de manera sostenida, la IA enviará una instrucción directa al **PLC (Controlador Lógico Programable)** de la máquina mediante protocolos industriales (como Modbus o OPC UA) para:
  1. Reducir la velocidad/carga de la máquina de forma automática para disminuir la fricción y enfriar el motor.
  2. Si el riesgo persiste, ejecutar una parada segura controlada para evitar daños catastróficos o accidentes laborales en la planta, todo esto en fracciones de segundo.
